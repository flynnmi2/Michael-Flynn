INTRODUCTION:


Methicillin-resistant Staphylococcus Aureaus, commonly known as MRSA, is a major problem in today's society. It is a bacterium that is one of the major causes in infection-related deaths because of its resistance to antibiotics. Beta-lactams like penicillins are the main antibiotic that targets the gram-positive bacteria,however in todays society, there is a problem with antibiotics and their lessening strength to fight against infections. The beta-lactams work by targeting the cell walls of the cells, however MRSA has a way of avoiding this. To begin, before the bacteria is considered methicillin-resistant, it is first methicillin-susceptible(MSSA). MSSA is killed by antibiotics because it lacks the DNA to code for a protein called mec-A protein which breaks down the antibiotic before it becomes effective. In order for MSSA to become MRSA, it must obviously obtain the DNA. This is done through a multitude of ways-mutation for example-but it is most often picked up by a MSSA from an external plasmid that was from a previous MRSA cell. with plasmid transer, the MSSA can now code for the mec-A protein and thus stop antibiotics from working. This is the consequential problem. Staphylococcus Aureaus has created a way to block death and trive in our bodies. This is what I want to dive into in this project. It seems that the reason Staphylococcus Aureaus has been able to do this by pass is from some original mutation to create the mec-A protein. This mutation must have started from exposure to the antibiotics themselves which led to MSSA turning into MRSA. Now, MRSA is an extremely dangerous bacteria often leading to death because almost nothing can stop it. My focus for this project is what does MRSA antibiotic resistance look like currently, and from this I would like to see what it may look like in the future. I would like to try and find a correlation between prescribed antibiotics and MRSA antibiotic resistance because I believe this is the root cause of why it has become so disasterous-the fact being that if people do not effectively completely kill the bacteria from their body it has the opportuntiy to mutate into MRSA and cause disaster. I think that the rise in MRSA infection concern is stemmed from the fact that antibiotics are not taken as seriously as they need to be and this is what is the result from not using antibiotics wisely. What will the future look like? Do we stand a chance against this tiny opponent?
